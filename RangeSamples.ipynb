# trycky important range examples 


```python
#start == stop â†’ nothing generated
#it indicates the empty 
print(list(range(5,5)))

#Default step confusion
print(list(range(5, -1)))
print(list(range(5, 1)))

#print(range(1,10,0)) # ValueError

#Negative step (reverse loop)
print(list(range(10, 0, -3)))


#Membership testing (VERY IMPORTANT)
10 in range(20)
if 15 in range (30):
    print('value of ',True)
    
elif():
    print('value of ',False)
    
#range equality (big trap)
print(range(0, 10) == range(0, 10, 1))

#Different arguments, same sequence

print(range(0, 10, 2) == range(0, 11, 2))

#nagetive indexing 

ls=list(range(20))
print(ls[3])
print(ls[-3])

r=range(120)
print(r[-10])

#index access with step

r =range(2,20,3)
print('kk',r)
for i in r:
    print(i)
    
    
print(r[-2])

r = range(2, 20, 3)
print(r[2])
```
