```python
# substring search
#diffrence between replace and substring search
# substring search is to find the position of a substring within a string, while replace is to replace occurrences of a substring with another substring.
# example:
s = "hello world"
substring = "world"
position = s.find(substring)
if position != -1:
    print(f"Substring '{substring}' found at position {position}.")
else:
    print(f"Substring '{substring}' not found.")    
    
# example of substring 
s = "hello ravi how are you and ravi is a good boy"
substring = "ravi"
count = s.count(substring)

print(f"The substring '{substring}' appears {count} times in the string.")  
```

```python
# make translation table
# difference between maketrans and translate
# maketrans is used to create a translation table that maps characters to their replacements, while translate is used to apply the translation table to a string, replacing characters according to the mappings defined in
 # example:
# Create a translation table that maps 'a' to '1', 'e' to '2', and 'i' to '3'
translation_table = str.maketrans('aei', '123')
# Apply the translation table to a string
original_string = "hello world"
translated_string = original_string.translate(translation_table)
print(translated_string)  # Output: "h2llo world"


x="hes a good boy"
y="rk"
rs=x.maketrans("hes","123")
print(x.translate(rs))
```

```python
text="hello world"
table=str.maketrans("","","lo")
result=text.translate(table)
print(result) 
# explain the above code
# In the above code, we are creating a translation table using the `str.maketrans` method. The first argument is an empty string, which means we are not specifying any characters to be replaced. The second argument is also an empty string, which means we are not specifying any characters to replace with. The third argument is the string "lo", which means we want to delete the characters 'l' and 'o' from the original string. The `translate` method is then used to apply this translation table to the original string "hello world", resulting in the output "he wrd".     
```

```python
s="ravi is good boy"
rs=s.maketrans(" "," ","goo") # In this line, we are creating a translation table using the `str.maketrans` method. The first argument is a space character, which means we want to replace spaces with something else. The second argument is also a space character, which means we want to replace spaces with spaces (i.e., keep them unchanged). The third argument is the string "goo", which means we want to delete the characters 'g', 'o', and 'o' from the original string. The resulting translation table will be used to remove these characters from the string when we call the `translate` method.
r=s.translate(rs)
print(r)
```
