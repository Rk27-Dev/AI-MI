```python
#walrus operator
# The walrus operator allows you to assign values to variables as part of an expression.
# It is represented by the symbol := and was introduced in Python 3.8.
# This operator is particularly useful in situations where you want to both assign a value to a variable

# and use that value within a larger expression, such as in loops or conditional statements.
n = 10
if (square := n * n) > 50:
    print(f"The square of {n} is {square}, which is greater than 50.")
else:
    print(f"The square of {n} is {square}, which is not greater than 50.")      
```

```python
# sum of the degits 
num = 123454
sum_of_digits = 0   
while num > 0:
    digit = num %  10  # Get the last digit 
    print("digit:",digit)
  
    sum_of_digits += digit  # Add it to the sum
    print("sum_of_digits:",sum_of_digits)
    
    num //= 10  # Remove the last digit
    #value of num after removing last digit
    print("num after removing last digit:",num) 
print("Sum of digits:", sum_of_digits)
```
