```python
# static  variable

class Student :
    school_name="ABC School"
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def talk(self):
       
        self.mark=90
        print(f"Student Name: {self.name.upper()}, Age: {self.age }, Mark: {self.mark}, School: {Student.school_name}")
    # static method
    @staticmethod
    def get_school_name():
        return Student.school_name  
s1=Student("ravi", 20)
s1.talk()
print(Student.get_school_name())
```
